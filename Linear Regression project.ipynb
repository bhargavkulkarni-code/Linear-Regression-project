```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
fmcg=pd.read_csv('C:\\Users\\bharg\\Desktop\\Bhargav\ExportSales.csv')
fmcg
```

```python
fmcg=fmcg.rename(columns={'Destenation':"Destination"})
fmcg
```

```python
fmcg.duplicated().sum()
```

```python
fmcg.isnull().sum()
```

```python
numerical=[fea for fea in fmcg.columns if fmcg[fea].dtypes!=object]
numerical
```

```python
categorical=[fea for fea in fmcg.columns if fmcg[fea].dtypes==object]
categorical
```

```python
# fmcg data can be classified in to 2 types one is numerical data and another is categorical data
```

```python
fmcgnumerical=fmcg[numerical]
fmcgnumerical
```

```python
# fmcg numerical data doesnot have null values
```

```python
fmcgcategorical=fmcg[categorical]
fmcgcategorical
```

```python
fmcgcategorical=fmcgcategorical.fillna(fmcgcategorical.mode().iloc[0])
fmcgcategorical
```

```python
# fmcg categorical null data  filled with mode
```

```python
fmcgnumerical.isnull().sum()
```

```python
fmcgcategorical.isnull().sum()
```

```python
fmcgcategorical['Category']=fmcgcategorical['Category'].replace('Category1',1).replace('Category2',2).replace('Category3',3).replace('Category4',4).replace('Category5',5).replace('Category6',6).replace('Category7',7).replace('Category8',8).replace('Category9',9).replace('Category10',10)
fmcgcategorical['Category']=fmcgcategorical['Category'].astype('int64')
```

```python
n=fmcgnumerical['QTY']
```

```python
x=pd.concat([n,fmcgcategorical['Category']],axis=1)
x
```

```python
y=fmcgnumerical['Total Price']
y
```

```python
# x is independent variable
```

```python
# y is dependent variable
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=12)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr
```

```python
lr.fit(x_train,y_train)
```

```python
y_prediction=lr.predict(x_test)
y_prediction
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mse=mean_squared_error(y_test,y_prediction)
r_squared=r2_score(y_test,y_prediction)
r_squared
```

```python
from math import sqrt
rmse=sqrt(mse)
print('mean squared error :',mse)
print('mean squared error :',rmse)
print('mean squared error :',r_squared)
```

```python
combined_actual_and_predicted_data=pd.DataFrame({'actual':y_test,'predicted':y_prediction})
combined_actual_and_predicted_data
```
